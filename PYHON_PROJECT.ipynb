```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_excel('myexcel.xlsx')

# Replace 'Height' column with random values between 150 and 180
df['Height'] = np.random.randint(150, 181, size=len(df))

# Save the cleaned data
df.to_excel('cleaned_employee_data.xlsx', index=False)
```

```python
# Calculate distribution
team_distribution = df['Team'].value_counts()

# Calculate percentages
total_employees = len(df)
team_percentages = (team_distribution / total_employees) * 100

print("Team Distribution:")
print(team_distribution)
print("\nTeam Percentages:")
print(team_percentages)
```

```python
position_distribution = df['Position'].value_counts()

print("Position Distribution:")
print(position_distribution)
```

```python
# Calculate age distribution
age_distribution = df['Age'].value_counts().sort_index()

print("Age Distribution:")
print(age_distribution)

# Find the predominant age group
predominant_age = age_distribution.idxmax()
print(f"\nPredominant Age Group: {predominant_age}")
```

```python
# Group by Team and Position, then sum the salaries
team_position_salary = df.groupby(['Team', 'Position'])['Salary'].sum().reset_index()

# Find the team and position with the highest salary expenditure
max_salary_row = team_position_salary.loc[team_position_salary['Salary'].idxmax()]

print(f"Highest Salary Expenditure: Team '{max_salary_row['Team']}', Position '{max_salary_row['Position']}'")
```

```python
# Calculate correlation
correlation = df['Age'].corr(df['Salary'])
print(f"Correlation between Age and Salary: {correlation}")

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Salary', data=df)
plt.title('Correlation between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Team', data=df)
plt.title('Distribution of Employees Across Teams')
plt.xlabel('Team')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Position', data=df)
plt.title('Distribution of Employees by Position')
plt.xlabel('Position')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], kde=True)
plt.title('Age Distribution of Employees')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```
